```r
numberguessing <- function(){
  number <- sample(0:100, 1)
  remaining= 15
  message("Try to guess the number I invented! (Between 0 and 100)\n")
  while(remaining > 0){
    message("Remaining tries: ", remaining)
    guessed <- readline(prompt="Your guess: ")
    if(guessed == number){
      break
    } else {
      again <- readline(prompt="Wrong! Guess again? (y/n)\n")
      if(!(again %in% c('y', 'n'))){
        again <- 'y'
      }
      if(again == 'n'){
        break
      }
      remaining = remaining -1
    }
  }
  if(remaining == 0 || again == 'n'){
    cat("Game over! My number was: ", number)
  } else {
    cat("Congratulations! You guessed right: ", number)
  }
}
```

```r
numberguessing()
```

```r
summary(esoph)
print('Cases by age')
print(rowsum(esoph$ncases, esoph$agegp))
print('Cases by alcohol consumption')
print(rowsum(esoph$ncases, esoph$alcgp))
print('Cases by tobacco consumption')
print(rowsum(esoph$ncases, esoph$tobgp))
print('Controls by age')
print(rowsum(esoph$ncontrols, esoph$agegp))
```

```r
plot(unique(esoph$agegp),rowsum(esoph$ncases, esoph$agegp), col="blue", main="Age vs cases", xlab="age group", ylab = "cases", type="b")
plot(unique(esoph$alcgp),rowsum(esoph$ncases, esoph$alcgp), col="green", main="Alcohol vs cases", xlab="Alcohol consumption", ylab = "cases")
plot(unique(esoph$tobgp),rowsum(esoph$ncases, esoph$tobgp), col="red", main="Tobacco vs cases", xlab="Tobacco consumption", ylab = "cases")
```

Interestingly enough, the number of cases of esophageal cancer seems to be mostly influenced by age, as more alcohol consumption does not consistently affect the number of cases. Neither does tobacco consumption, the actual advice from this data would actually be to smoke more, as there are less cases in people who smoke more.
